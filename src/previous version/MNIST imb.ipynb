```python
# new metrics
import numpy as np
from keras import objectives
from keras import backend as K
import theano
 
import theano.tensor as T
def mcos(y_true, y_pred):
    cost=K.constant(0)
    for k in range(10):
        d=K.log(1+K.exp((int(bool(y_true[k]==K.min(y_true)))*2-1)*(y_pred[k]-y_true[k])))
        cost=cost+d
    return cost
 
```
